# Bidirectional RNNs for Text Sentiment Classification
A bidirectional RNN acts similar to a RNN where it captures sequence information in order. As its name suggests, its difference lies in using two regular RNNs and processing the input sequence chronologically and antichronologically respectively, and then combining their representations. Processing sequential data in both dimensions allow the network to catch additional patterns which may have been overlooked by a single unidirectional RNN. 

Bidirectional RNNs are popular for NLP problems since the context of words in a sentence isn't dependent on its position in the sentence. As such, we will be testing how bidirectional RNNs perform on text sentiment classification using IMDB reviews.

```python
from keras.datasets import imdb
from keras.preprocessing import sequence
from keras import models, layers
```

```python
max_features = 10000
max_len = 500

(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=max_features)

# reverse sequences
x_train = [x[::-1] for x in x_train]
x_test = [x[::-1] for x in x_test]

# Padding
x_train = sequence.pad_sequences(x_train, maxlen=max_len)
x_test = sequence.pad_sequences(x_test, maxlen=max_len)
```

## Unidirectional LSTM with reversed sequences

```python
model = models.Sequential()
model.add(layers.Embedding(max_features, 128))
model.add(layers.LSTM(32))
model.add(layers.Dense(1, activation='sigmoid'))

model.compile(
    optimizer='rmsprop',
    loss='binary_crossentropy',
    metrics=['acc']
)

history = model.fit(
    x_train,
    y_train,
    epochs=10,
    batch_size=128,
    validation_split=0.2
)
```

```python
import matplotlib.pyplot as plt

# Plot results
acc = history.history['acc']
val_acc = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(1, len(acc)+1)

f, ax = plt.subplots(1,2, figsize=(14,4))

ax[0].plot(epochs, acc, 'bo', label='Training acc')
ax[0].plot(epochs, val_acc, 'r', label='Validation acc')
ax[0].set_title('Training and validation accuracy')
ax[0].legend()

ax[1].plot(epochs, loss, 'bo', label='Training loss')
ax[1].plot(epochs, val_loss, 'r', label='Validation loss')
ax[1].set_title('Training and validation loss')
ax[1].legend()

plt.show()
```

## Bidirectional LSTM

```python
bi_model = models.Sequential()
bi_model.add(layers.Embedding(max_features, 32))
bi_model.add(layers.Bidirectional(layers.LSTM(32, dropout=0.1, implementation=2)))
bi_model.add(layers.Dense(1, activation='sigmoid'))

bi_model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['acc']
)

history = bi_model.fit(
    x_train,
    y_train,
    epochs=20,
    batch_size=128,
    validation_split=0.2
)
```

```python
# Plot results
acc = history.history['acc']
val_acc = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(1, len(acc)+1)

f, ax = plt.subplots(1,2, figsize=(14,4))

ax[0].plot(epochs, acc, 'bo', label='Training acc')
ax[0].plot(epochs, val_acc, 'r', label='Validation acc')
ax[0].set_title('Training and validation accuracy')
ax[0].legend()

ax[1].plot(epochs, loss, 'bo', label='Training loss')
ax[1].plot(epochs, val_loss, 'r', label='Validation loss')
ax[1].set_title('Training and validation loss')
ax[1].legend()

plt.show()
```

We begin fitting after around 5 epochs, so lets train the same model for less epochs.

```python
bi_model_es = models.Sequential()
bi_model_es.add(layers.Embedding(max_features, 32))
bi_model_es.add(layers.Bidirectional(layers.LSTM(32, dropout=0.1, implementation=2)))
bi_model_es.add(layers.Dense(1, activation='sigmoid'))

bi_model_es.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['acc']
)

history = bi_model_es.fit(
    x_train,
    y_train,
    epochs=5,
    batch_size=128,
    validation_split=0.2
)
```

```python
# Plot results
acc = history.history['acc']
val_acc = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(1, len(acc)+1)

f, ax = plt.subplots(1,2, figsize=(14,4))

ax[0].plot(epochs, acc, 'bo', label='Training acc')
ax[0].plot(epochs, val_acc, 'r', label='Validation acc')
ax[0].set_title('Training and validation accuracy')
ax[0].legend()

ax[1].plot(epochs, loss, 'bo', label='Training loss')
ax[1].plot(epochs, val_loss, 'r', label='Validation loss')
ax[1].set_title('Training and validation loss')
ax[1].legend()

plt.show()
```

```python
# Evaluate on Test set
testLoss, testAcc = bi_model_es.evaluate(x_test, y_test)

print('test loss', testLoss)
print('test accuracy', testAcc)
```
